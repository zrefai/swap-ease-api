```python
import pandas

df = pandas.read_csv('../data/DOODLES.csv', index_col="token_id")
df.shape
```

```python
import umap.plot

reducer = umap.UMAP().fit(df)
umap.plot.connectivity(reducer)
```

```python
import matplotlib.pyplot as plt

def draw_umap(data, n_neighbors=15, min_dist=0.1, metric='euclidean', title=''):
    fit = umap.UMAP(
        n_neighbors=n_neighbors,
        min_dist=min_dist,
        metric=metric
    )
    u = fit.fit_transform(data);
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.scatter(u[:,0], u[:,1])

    plt.title(title, fontsize=18)

for d in (0.0, 0.1, 0.25, 0.5, 0.8, 0.99):
    draw_umap(df, min_dist=d, title='min_dist = {}'.format(d))
```
